Task - 3 - Outlier Detection:
Tasks:
1. Identify outlier countries in South Asia based on their Score and GDP per Capita.
2. Define outliers using the 1.5 × IQR rule.
3. Create a scatter plot with GDP per Capita on the x-axis and Score on the y-axis, highlighting outliers
in a different color.
4. Discuss the characteristics of these outliers and their potential impact on regional averages.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the filtered South Asia dataset
filtered_file_path = 'Filtered_South_Asia_Happiness.csv'
filtered_df = pd.read_csv(filtered_file_path)

# Task 3: Detecting Outliers Using the 1.5 * IQR Method

# Calculate the Interquartile Range (IQR) for 'score' and 'Log GDP per capita'
score_q1 = filtered_df['score'].quantile(0.25)
score_q3 = filtered_df['score'].quantile(0.75)
score_iqr = score_q3 - score_q1

gdp_q1 = filtered_df['Log GDP per capita'].quantile(0.25)
gdp_q3 = filtered_df['Log GDP per capita'].quantile(0.75)
gdp_iqr = gdp_q3 - gdp_q1

# Establish outlier boundaries
score_min_threshold = score_q1 - 1.5 * score_iqr
score_max_threshold = score_q3 + 1.5 * score_iqr

gdp_min_threshold = gdp_q1 - 1.5 * gdp_iqr
gdp_max_threshold = gdp_q3 + 1.5 * gdp_iqr

# Filter outliers
outlier_records = filtered_df[
    (filtered_df['score'] < score_min_threshold) | (filtered_df['score'] > score_max_threshold) |
    (filtered_df['Log GDP per capita'] < gdp_min_threshold) | (filtered_df['Log GDP per capita'] > gdp_max_threshold)
]

# Display the detected outliers
print("Detected Outliers Based on Score and GDP per Capita:")
print(outlier_records)

# Visualize the outliers using a scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_df['Log GDP per capita'], filtered_df['score'], color='skyblue', label='Normal Data')
plt.scatter(outlier_records['Log GDP per capita'], outlier_records['score'], color='red', label='Outliers')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.title('Outlier Detection in South Asia (GDP vs Score)')
plt.legend()
plt.grid(True)
plt.show()

```

6. Outlier Detection:

• Identify outlier countries in both regions based on Score and GDP per Capita.

• Plot these outliers and discuss their implications.


```python
# Task 6: Outlier Detection

south_asia_df = pd.read_csv('South_Asia_Dataset.csv')
middle_east_df = pd.read_csv('Middle_East_Dataset.csv')


# Function to detect outliers using the 1.5 * IQR rule
def detect_outliers(df, metric):
    q1 = df[metric].quantile(0.25)
    q3 = df[metric].quantile(0.75)
    iqr = q3 - q1

    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    outliers = df[(df[metric] < lower_bound) | (df[metric] > upper_bound)]
    return outliers

# Detect outliers in both datasets
south_asia_outliers = detect_outliers(south_asia_df, 'score')
middle_east_outliers = detect_outliers(middle_east_df, 'score')

# Print outliers
print("South Asia Outliers based on Score:")
print(south_asia_outliers[['Country name', 'score', 'Log GDP per capita']])

print("\nMiddle East Outliers based on Score:")
print(middle_east_outliers[['Country name', 'score', 'Log GDP per capita']])

# Visualize the outliers using scatter plots
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 8))

# South Asia Scatter Plot
plt.subplot(2, 1, 1)
plt.scatter(south_asia_df['Log GDP per capita'], south_asia_df['score'], color='skyblue', label='Normal')
plt.scatter(south_asia_outliers['Log GDP per capita'], south_asia_outliers['score'], color='red', label='Outliers')
plt.title('South Asia: GDP vs Score (Outliers Highlighted)')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid(True)

# Middle East Scatter Plot
plt.subplot(2, 1, 2)
plt.scatter(middle_east_df['Log GDP per capita'], middle_east_df['score'], color='lightgreen', label='Normal')
plt.scatter(middle_east_outliers['Log GDP per capita'], middle_east_outliers['score'], color='red', label='Outliers')
plt.title('Middle East: GDP vs Score (Outliers Highlighted)')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Task 7: Visualization - Boxplots for Score Comparison

south_asia_df = pd.read_csv('South_Asia_Dataset.csv')
middle_east_df = pd.read_csv('Middle_East_Dataset.csv')

import matplotlib.pyplot as plt

# Create boxplots comparing the Score distributions
plt.figure(figsize=(8, 6))

# Boxplot creation
plt.boxplot([south_asia_df['score'], middle_east_df['score']],
            labels=['South Asia', 'Middle East'], patch_artist=True,
            boxprops=dict(facecolor='skyblue', color='blue'),
            medianprops=dict(color='red'), whiskerprops=dict(color='blue'))

# Plot settings
plt.title('Comparison of Score Distributions between South Asia and Middle East')
plt.ylabel('Score')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the plot
plt.show()
```
